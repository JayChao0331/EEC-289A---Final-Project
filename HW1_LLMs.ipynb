# Question 7: Replaceing the exp(.) operation with the simple abs(.) operation in both numerator and denominator of the Softmax function within the original model from Question 1.

```python
!python train.py config/train_shakespeare_char.py --use_abs_softmax=True
```

### Compared with the results of the original model of Question 1:

Interestingly, we have a higher running time, along with higher training and final iteration loss values, when switching to using absolute value instead of exponential for the softmax function. However, the validation loss is slightly lower when compared to Question 1's original model. This is due to the fact that custom softmax is being implemented in the manual version of attention and thus flash attention is disabled when using custom softmax. This is the likely reason for this slowdown. However, if we tried to use the original softmax function with the same manual version of non-flash attention, the running time may very well be higher than the current time of 16408.14ms. In the future, the flash version of attention, specifically within the scaled_dot_product_attention() function, could also be modified to support the use of custom softmax functions.
